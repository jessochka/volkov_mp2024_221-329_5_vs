## Множество (Set)

```
// HashSet - неупорядоченное множество
// SortedSet - упорядоченное множество
```

```
HashSet<int> numbers = new();
numbers.Add(1);
numbers.Add(2);
numbers.Add(2);
numbers.Add(2);
numbers.Add(3);
numbers
```

```
// Проверка наличия элемента
Console.WriteLine(numbers.Contains(1));
Console.WriteLine(numbers.Contains(4));
```
